```python
import random

def tas_kagit_makas_buz_abdulkadir_kocyigit():
    """
    Taş, Kağıt, Makas, Buz oyununu başlatan fonksiyon.
    
    Kullanıcıdan bir seçim yapması istenir ve bilgisayar da rastgele bir seçim yapar.
    İki galibiyete ulaşan taraf oyunu kazanır. Kullanıcı oyundan çıkmayı da tercih edebilir.

    """
    
    # Oyun başlangıç mesajları
    print("Taş, Kağıt, Makas, Buz Oyununa Hoşgeldiniz!")
    print("Oyunu oynamak için 'taş', 'kağıt', 'makas' veya 'buz' yazın. "
          "Oyundan çıkmak için 'çıkış' yazabilirsiniz.")
    print("Unutmayın: Taş buzu yener, taş makası yener; kağıt taşı yener; "
          "makas kağıdı yener; makas buzu yener; ve buz kağıdı yener.")
    print("Seçimlerinizi yaparken bunları dikkate almayı unutmayın.")
    print("İYİ OYUNLAR VE BOL ŞANSLAR")
    
    # Oyun seçenekleri
    secenekler = ["taş", "kağıt", "makas", "buz"]
    
    # Oynanan oyun ve tur sayısı sayaçları
    oyun_sayisi = 0
    tur_sayisi = 0
    
    while True:
        # Galibiyet sayılarının başlangıç değeri
        oyuncu_galibiyetleri = 0
        bilgisayar_galibiyetleri = 0
        
        # Oyun başına tur sayısını sıfırla
        oyun_sayisi += 1
        print(f"Oyun {oyun_sayisi} başlıyor...")
        
        # Oyuncu veya bilgisayar 2 galibiyete ulaşana kadar oyun devam eder
        while oyuncu_galibiyetleri < 2 and bilgisayar_galibiyetleri < 2:
            # Kullanıcıdan seçim alınır
            oyuncu_secimi = input("Seçiminizi yapın (taş, kağıt, makas, buz): ").lower()
            
            # Kullanıcı oyundan çıkmak isterse
            if oyuncu_secimi == 'çıkış':
                print("Oyundan çıkılıyor...")
                return
            
            # Geçersiz seçim durumu
            if oyuncu_secimi not in secenekler:
                print("Geçersiz bir seçim yaptınız, lütfen tekrar deneyin.")
                continue
            
            # Bilgisayarın rastgele seçimi
            bilgisayar_secimi = random.choice(secenekler)
            print(f"Bilgisayarın seçimi: {bilgisayar_secimi}")
            
            # Oynanan tur sayısını artır
            tur_sayisi += 1
            
            # Oyun kurallarına göre galip belirlenir
            if oyuncu_secimi == bilgisayar_secimi:
                print("Berabere!")
            elif (
                (oyuncu_secimi == "taş" and bilgisayar_secimi in ["buz", "makas"]) or
                (oyuncu_secimi == "kağıt" and bilgisayar_secimi == "taş") or
                (oyuncu_secimi == "makas" and bilgisayar_secimi in ["kağıt", "buz"]) or
                (oyuncu_secimi == "buz" and bilgisayar_secimi == "kağıt")
            ):
                print("Bu turu kazandınız!")
                oyuncu_galibiyetleri += 1
            else:
                print("Bu turu bilgisayar kazandı!")
                bilgisayar_galibiyetleri += 1
            
            # Skorun ekrana yazdırılması
            print(f"Skor: Oyuncu {oyuncu_galibiyetleri} - Bilgisayar {bilgisayar_galibiyetleri}")
            print(f"Toplam oynanan tur sayısı: {tur_sayisi}")
        
        # Oyunun galibini belirle ve mesajı ekrana yazdır
        if oyuncu_galibiyetleri == 2:
            print("Tebrikler! Oyunu kazandınız!")
        else:
            print("Bilgisayar oyunu kazandı!")
        
        # Kullanıcıya yeni oyun isteyip istemediğini sor
        devam = input("Başka bir oyun oynamak ister misiniz? (evet/hayır): ").lower()
        if devam != 'evet':
            print("Oyunu sonlandırdınız. Teşekkürler!")
            break
        
        # Bilgisayarın oyuna devam etmek isteyip istemediği belirlenir
        bilgisayar_devam = random.choice(['evet', 'hayır'])
        print(f"Bilgisayarın devam etme isteği: {bilgisayar_devam}")
        if bilgisayar_devam != 'evet':
            print("Bilgisayar oyuna devam etmek istemiyor. Oyun sona erdi.")
            break
    
    # Oyun sonu mesajları
    print(f"Toplam oynanan oyun sayısı: {oyun_sayisi}")
    print(f"Toplam oynanan tur sayısı: {tur_sayisi}")
```

```python
# Fonksiyonu çalıştırma
tas_kagit_makas_buz_abdulkadir_kocyigit()
```
